```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
train = pd.read_csv("train.csv")
```

```python
train.head()
```

# Simple Logistic Regression

```python
X = train.drop(["Exited", "id", "CustomerId", "Surname"], axis=1)
Y = train.Exited
```

```python
X_train = pd.get_dummies(X, columns=["Gender", "Geography"], drop_first=True)
X_train.head()
```

```python
model = LogisticRegression()
model.fit(X_train, Y)
```

```python
test = pd.read_csv("test.csv")
X_id = test["id"]
X_test = test.drop(["id", "CustomerId", "Surname"], axis=1)
X_test = pd.get_dummies(X_test, columns=["Gender", "Geography"], drop_first=True)
X_test.head()
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
submission = pd.DataFrame({"id": X_id, "Exited": predictions}).set_index("id")
submission.head()
```

```python
submission.to_csv("submission.csv")
```
